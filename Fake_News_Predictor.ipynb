```python
import nltk.corpus
nltk.download('stopwords')
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv('/content/Fake News Predictor.csv')
df.head()
```

```python
df.dropna(inplace=True) # Drop null values
df.isnull().sum() # Calculate the sum of null values
```

```python
 df['content'] = df['title'] + ''+ df['author']
 df
```

```python
# Seperating the data and label
```

```python
x = df.drop(columns='label', axis=1)
y = df['label']
```

```python
print(x)
```

```python
print(y)
```

```python
port_stem = PorterStemmer()
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content
```

```python
df['content'] = df['content'].apply(stemming)
df
```

```python
# Seperating the columns and data

x = df['content'].values
y = df['label'].values

print(x)
```

```python
print(y)
```

```python
y.shape
```

```python
# To convert texts into numbers, use Tfidfvectorizer

vectorizer = TfidfVectorizer()
vectorizer.fit(x)

x = vectorizer.transform(x)
```

```python
print(x)
```

```python
# Splitting the data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
#Define the model

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
#Train the model

model.fit(x_train, y_train)
```

```python
#Test the model

#Check the training data accuracy

from sklearn.metrics import accuracy_score
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print('Accuracy score of the training data : ', training_data_accuracy)
```

```python
#Check the test data accuracy which is significant
pred = model.predict(x_test)
pred
```

```python
pred_check = accuracy_score(pred, y_test)
pred_check
```

```python
#Predictive Analysis


input = x_test[1]
prediction = model.predict(input)
print(prediction)


if (prediction[0]==0):
  print('The news is Real')
else:
  print('The news is Fake')
```

```python
# prompt: Next steps in this code to determine if any given text is fake or real news?

def predict_news(news_text):
  """Predicts whether a given news text is real or fake.

  Args:
    news_text: The text of the news article to be classified.

  Returns:
    A string indicating whether the news is real or fake.
  """

  stemmed_news = stemming(news_text)
  vectorized_news = vectorizer.transform([stemmed_news])
  prediction = model.predict(vectorized_news)

  if prediction[0] == 0:
    return 'The news is Real'
  else:
    return 'The news is Fake'

# Example usage:
new_news_text = "Your news text here"
prediction_result = predict_news(new_news_text)
print(prediction_result)
```
